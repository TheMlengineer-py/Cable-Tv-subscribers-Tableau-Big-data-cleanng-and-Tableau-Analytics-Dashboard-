# CableTVSubscribersData preprocessing

```python
#Read CSV File
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_csv('CableTVSubscribersData.csv')
Cable_tv_subscribers=df
Cable_tv_subscribers
```

```python
#create function to check for missing values in the data

def count_of_null(Cable_tv_subscribers):
    count=Cable_tv_subscribers.isnull().sum().sum()
    return count

# print missing value count in the data
count_null = count_of_null(Cable_tv_subscribers)
print(count_null)
```

```python
#create function to check duplicate values in the data
def check_duplicates(Cable_tv_subscribers):
    count=Cable_tv_subscribers.duplicated().sum().sum()
    return count

# print duplicate values in the data
count_duplicates = check_duplicates(Cable_tv_subscribers)
print(count_duplicates)
```

```python
#create a function to check information about the data
def about_data (Cable_tv_subscribers):
        about=Cable_tv_subscribers.info()
        return about
    
#print information about the data
info=about_data (Cable_tv_subscribers)
print(info)
```

```python
# Convert the age column to integer
Cable_tv_subscribers['age'] = Cable_tv_subscribers['age'].astype(int)

# Print the updated DataFrame
Cable_tv_subscribers
```

```python
# create function to print unique values in the dataframe

def print_unique_values(Cable_tv_subscribers):
    for column in Cable_tv_subscribers.columns:
        unique_values = Cable_tv_subscribers[column].unique()
        print(f"Column '{column}' has {len(unique_values)} unique values:")
        print(unique_values)
print_unique_values(Cable_tv_subscribers)
```

```python
#Create function to check statistical summary of the dataset
def stat_summary (Cable_tv_subscribers):
    statistics_sum = Cable_tv_subscribers.describe()
    return statistics_sum
#print statistical summary of data
summary = stat_summary (Cable_tv_subscribers)
summary
```

```python
#create function to find columns with Categorical Values

def categorcal_columns (Cable_tv_subscribers):
    cat_cols= Cable_tv_subscribers.select_dtypes(exclude=['int','float']).columns.tolist()
    return cat_cols
#print categorical columns
categorical = categorcal_columns (Cable_tv_subscribers)
categorical
```

```python
#create function to find columns with Numerical Values

def numerical_columns (Cable_tv_subscribers):
    num_cols = Cable_tv_subscribers.select_dtypes(include=['int','float']).columns.tolist()
    return num_cols
#print numerical columns
numerical = numerical_columns (Cable_tv_subscribers)
numerical
```

```python
#Removing extra spaces from the features Using strip()
Cable_tv_subscribers.columns = Cable_tv_subscribers.columns.str.strip()
```

```python
#Finding Outliers using IQR score

#The IQR is the first quartile subtracted from the third quartile; 
#these quartiles can be clearly seen on a box plot on the data.
# calculate IQR score
Q1 = Cable_tv_subscribers.quantile(0.25)
Q3 = Cable_tv_subscribers.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
# IQR of each column
```

```python
#remove outlier
import warnings
warnings.simplefilter(action="ignore", category=FutureWarning)

Cable_tv_subscribers = Cable_tv_subscribers[~((Cable_tv_subscribers < (Q1 - 1.5 * IQR)) |(Cable_tv_subscribers > (Q3 + 1.5 * IQR))).any(axis=1)]
Cable_tv_subscribers.shape
```

```python
# load cleaned Cable_tv_subscribers dataframe into CSV file.
Cable_tv_subscribers.to_csv('Fixed_Cable_tv_subscribers_data.csv', index=False, header=True)
```

# Churn Data preprocessing 

```python
df=pd.read_csv('churned_clean_.csv')
churn=df
churn
```

```python
#create function to check for missing values in the data

def count_of_null(churn):
    count=churn.isnull().sum().sum()
    return count

# print missing value count in the data
count_null = count_of_null(churn)
print(count_null)
```

```python
#create function to check duplicate values in the data
def check_duplicates(churn):
    count=churn.duplicated().sum().sum()
    return count

# print duplicate values in the data
count_duplicates = check_duplicates(churn)
print(count_duplicates)
```

```python
#create a function to check information about the data
def about_data (churn):
        about=churn.info()
        return about
    
#print information about the data
info=about_data (churn)
print(info)
```

```python
#Remove irrelevant columns

churn=churn.drop(columns=['Item1','Item2','Item3','Item4','Item5','Item6','Item7','Item8','Interaction',
                          'UID', 'City', 'County','TimeZone','CaseOrder','Zip', 'Unnamed: 0',])
churn
```

```python
# create function to print unique values in the dataframe

def print_unique_values(churn):
    for column in churn.columns:
        unique_values = churn[column].unique()
        print(f"Column '{column}' has {len(unique_values)} unique values:")
        print(unique_values)
print_unique_values(churn)
```

```python
#Create function to check statistical summary of the dataset
def stat_summary (churn):
    statistics_sum = churn.describe()
    return statistics_sum
#print statistical summary of data
summary = stat_summary (churn)
summary
```

```python
#create function to find columns with Categorical Values

def categorcal_columns (churn):
    cat_cols= churn.select_dtypes(exclude=['int','float']).columns.tolist()
    return cat_cols
#print categorical columns
categorical = categorcal_columns (churn)
categorical
```

```python
#create function to find columns with Numerical Values

def numerical_columns (churn):
    num_cols = churn.select_dtypes(include=['int','float']).columns.tolist()
    return num_cols
#print numerical columns
numerical = numerical_columns (churn)
numerical
```

```python
#Removing extra spaces from the features Using strip()
churn.columns = churn.columns.str.strip()
```

```python
#Finding Outliers using IQR score

#The IQR is the first quartile subtracted from the third quartile; 
#these quartiles can be clearly seen on a box plot on the data.
# calculate IQR score
#where Q3 is the 75th percentile of the data and Q1 is the 25th percentile of the data.
Q1 = churn.quantile(0.25)
Q3 = churn.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
# IQR of each column
```

```python
#remove outlier
import warnings
warnings.simplefilter(action="ignore", category=FutureWarning)
#Once the IQR is calculated, we can use it to identify outliers by defining a threshold range as follows;
#Lower threshold = Q1 - 1.5 * IQR
#Upper threshold = Q3 + 1.5 * IQR
#Any data points that fall outside of this range are considered outliers and removed from the dataset

churn = churn[~((churn < (Q1 - 1.5 * IQR)) |(churn > (Q3 + 1.5 * IQR))).any(axis=1)]
churn.shape
```

```python
# load cleaned churn dataframe into CSV file.
churn.to_csv('Fixed_churn.csv', index=False, header=True)
```
